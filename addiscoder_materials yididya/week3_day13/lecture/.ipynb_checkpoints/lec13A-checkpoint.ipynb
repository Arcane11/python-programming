As a reminder, the high level idea for quick sort is:
1. Choose a pivot element from the array
2. Partition the array into three subarrays, containing elements smaller than pivot, pivot itself, and elements larger than pivot
3. recursively quick sort last and first subarrays

Today, we will cover a different sorting algorithm which is also a divide and conquer algorithm like quicksort: mergesort.

## Divide and Conquer
Divide and conquer is a general, three-step pattern for algorithms:

1. Divide the given instance of the problem into several independent smaller instances of exactly the same problem
2. Delegate each smaller instance to the Recursion Fairy
3. Combine the solutions for the smaller instances into the final solution for the given instance

This should be similar to a lot of recursive functions we have seen already.

## Built-in Python Sort
